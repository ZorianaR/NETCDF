```python
import netCDF4 as nc
shp='GEAA-AEIv3.0M_SHP_2012.nc'
```

```python
shipping=nc.Dataset(shp)
```

```python
print("Count of objects: ",len(shipping.variables['gridID']))
```

## Якщо gridID=0, то lat=lon=0 і емісії повністю 0-ві, а якщо gridID!=0, то lat і lon мають вигляд *.**25 або *.**75

```python
count_not_zero_objects=0
count_zero_objects=0
for gridID, lat, lon, emis in zip(shipping.variables['gridID'],
                           shipping.variables['lat'],
                           shipping.variables['lon'],
                           shipping.groups['COMPOUNDS_EMISSION'].variables['CO2_']):
    if gridID==0 and lat==0 and lon==0 and set(emis)==set([0]):
        count_zero_objects+=1
    elif (gridID!=0 and (((round(float(lon),4)*10000)%100==75.0 or (round(float(lon),4)*10000)%100==25.0) 
        and ((round(float(lat),4)*10000)%100==75.0 or (round(float(lat),4)*10000)%100==25.0))):
        count_not_zero_objects+=1
    else:
        print("False")
        
print("Count of not zero-objects: ", count_not_zero_objects)
print("Count of zero-objects: ", count_zero_objects)
```

## Перевірка впорядкованості gridID по зростанню та визначення чи нема однакових gridID

```python
#створили список ненульвих gridID
not_zero_gridID=[]
for x in shipping.variables['gridID']:
    if x!=0:
        not_zero_gridID.append(float(x))

#Перевірка впорядкованості по зростанню та визначення чи нема однакових gridID
for i in range(len(not_zero_gridID)-1):
    if not_zero_gridID[i] < not_zero_gridID[i+1]:
        pass
    else:
        print('False')
        break
       
```

```python
for gridID, lat, lon, emis in zip(shipping.variables['gridID'],
                           shipping.variables['lat'],
                           shipping.variables['lon'],
                           shipping.groups['COMPOUNDS_EMISSION'].variables['CO2_']):
    if gridID==0:
        pass
    else:
        print('gridID={:<15}lon={:<20}lat={:<20}'.format(gridID,lon,lat))
        print('emission=',emis)
        print('\n\n')
```
